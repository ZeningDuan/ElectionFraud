```python
import pandas as pd
import time
import os
from urllib.parse import urlparse
import collections
from collections import Counter
```

```python
def dayslist(monthlengh):
    S = []
    l = list(range(1, monthlengh +1))
    s = [str(i) for i in l]
    for i in s:
        if int(i) < 10:
            ss = "0" + i     
        else:
            ss = i  
        S.append(ss)
   
    S.sort()
    return S


def datecheck(clean_date):
    years = []
    months = []
    days = []
    flag = ['Correct', 'Correct', 'Correct'] #Year, Month, Days

    for i in clean_date:
        years.append(i[:4])
        months.append(i[5:7])
        days.append(i[8:10])


    if len(set(years)) == 1:
        Year = str(list(set(years))[0])
        if Year == filename[-11 : -7]:
            print('Year is Correct!', 'Y: ', Year)
            
        else:
            msg = 'Error: Year does not match!'
            print(msg)
            flag[0] = msg
    else:
        msg = 'Error: More than one year info found!'
        print(msg)
        flag[0] = msg


    if len(set(months)) == 1:
        Month = str(list(set(months))[0])
        if Month == filename[-6 : -4]:
            print('Month is Correct!', 'M: ', Month)
            
            
        else:
        
            msg = 'Error: Month does not match!'
            print(msg)
            flag[1] = msg
            
    else:
        msg = 'Error: More than one month info!'
        print(msg)
        flag[1] = msg


    days_sort = list(set(days))
    days_sort.sort()
    if Month in ['01', '03', '05', '07', '08', '10', '12'] and len(list(set(days))) == 31:          
        if days_sort == dayslist(31):
            print('Days are Correct!')
            
        else:
            msg = 'Attention: Days are not completed!'
            print(msg)
            flag[2] = msg
            
    
    elif Month in ['04', '06', '09','11'] and len(list(set(days))) == 30:          
        if days_sort == dayslist(30):
            print('Days are Correct!')
            
        
        else:
            msg = 'Attention: Days are not completed!'
            print(msg)
            flag[2] = msg
        
    elif Month in ['02'] and len(list(set(days))) == 29:          
        if days_sort == dayslist(29):
            print('Days are Correct!')
            
        else:
            msg = 'Attention: Days are not completed!'
            print(msg)
            flag[2] = msg
    
    else:
        msg = days_sort
        print(msg)
        flag[2] = msg
        
    return flag
        

def checkdomains(URLs):
    domains = []
    for i in URLs:
        try:
            domain = urlparse(i).netloc
            domains.append(domain)
        except Exception as e:
            print(e)
            
    Domains = dict(Counter(domains))
    return Domains
```

```python
Sources = ['Foxnews', 'Breitbart',  'WashingtonPost', 'NYT', 'MSNBC', 'Huffpost']

record = {'Filename': [], '# of rows with empty date': [], 'Ratio of rows with empty date': [], 'Year Check': [], "Month Check": [], 'Day Check': [], 'Unique Domains': []}

for Source in Sources:
    path = "/Volumes/ZeDrive1/Projects/Covid_Bot Agenda setting/News Data/COVID-19_Media Cloud/" + Source + "_2020/"


    for root, dirs, files in os.walk(path):
        for filename in files:
            if Source in filename and '._' not in filename and 'data' not in filename and 'all' not in filename:

                print('File: ', filename)
                record['Filename'].append(filename)
                tic = time.time()

                data = pd.read_csv(path + filename)


                # ----Check Date ----
                # calculate the amount and ratio of rows with empty date information
                count_na = data["publish_date"].isna().sum()
                print("The amount of row with empty date is: ", count_na, "Ratio is: ", count_na/len(data))
                record['# of rows with empty date'].append(count_na)
                record['Ratio of rows with empty date'].append(count_na/len(data))

                #check if the date info is correct (correct if it matches with the target date shown in file name) 
                date = list(set(data['publish_date'])) 
                clean_date = [x for x in date if str(x) != 'nan']
                flag = datecheck(clean_date)
                record['Year Check'].append(flag[0])
                record['Month Check'].append(flag[1])
                record['Day Check'].append(flag[2])

                # ----Check URL domains ----
                URLs = list(data['url'])
                Domains = checkdomains(URLs)
                #Unique_Domains = list(set(Domains.keys()))
                record['Unique Domains'].append(Domains)


                print(Domains)
                print('-------------------------------------------')
                print('\n')

            
        
CheckResults = pd.DataFrame(record)
```

```python
CheckResults = CheckResults.sort_values(by=['Filename'])
```

```python
def sumdict(dflist):
    # printing initial dictionary
    print ("initial dictionary", str(dflist))
    
    # sum the values with same keys
    counter = collections.Counter()
    for d in dflist: 
        counter.update(d)
      
    result = dict(counter)
    
    print("resultant dictionary : ", result)
    
    return result
    
    
```
